```python
import pandas as pd


# データフレームの作成
df1 = pd.DataFrame({'ID': [1, 2],
                    '日本語': ['りんご', 'ぶどう'],
                    '英語': ['apple', 'grape']
})

df2 = pd.DataFrame({'ID': [3, 4],
                    '日本語': ['レモン', 'バナナ'],
                    '英語': ['lemon', 'banana']
})

# 縦結合
result = pd.concat([df1, df2], axis=0)

# 横結合
result = pd.concat([df1, df2], axis=1)

# 結果の表示
print(result)
```

```python


from openpyxl import Workbook
from openpyxl.styles import Alignment

# ワークブックとシートの作成
wb = Workbook()
ws = wb.active
ws.title = "請求書"

# 各種情報の準備
invoice_info = [
    ("B2", "請求書"),
    ("B4", "株式会社ABC"),
    ("B5", "〒101-0022 東京都千代田区神田練塀町300"),
    ("B6", "TEL:03-1234-5678 FAX:03-1234-5678"),
    ("B7", "担当者名:鈴木一郎 様"),
    ("F4", "No."),
    ("G4", "0001"),
    ("F5", "日付"),
    ("G5", "2025/10/02")
]

# 情報をループで書き込み
for i in range(len(invoice_info)):
    cell, value = invoice_info[i]
    ws[cell] = value

# ヘッダー行
header = ["商品名", "数量", "単価", "金額"]
header_pos = ["B10", "C10", "D10", "E10"]
for i in range(len(header)):
    ws[header_pos[i]] = header[i]

# 商品データ
products = [
    ("商品A", 2, 10000),
    ("商品B", 1, 15000)
]

# 商品情報の書き込み
start_row = 11
for i in range(len(products)):
    name, qty, price = products[i]
    amount = qty * price
    ws[f"B{start_row + i}"] = name
    ws[f"C{start_row + i}"] = qty
    ws[f"D{start_row + i}"] = price
    ws[f"E{start_row + i}"] = amount

# 金額計算
subtotal = sum([p[1] * p[2] for p in products])
tax = int(subtotal * 0.1)
total = subtotal + tax

# 金額欄
amounts = [("D15", "小計"), ("E15", subtotal),
           ("D16", "消費税"), ("E16", tax),
           ("D17", "合計"), ("E17", total)]

for i in range(0, len(amounts), 2):
    ws[amounts[i][0]] = amounts[i][1]
    ws[amounts[i+1][0]] = amounts[i+1][1]

# セルの中央揃え
center_cells = ["B2", "B10", "C10", "D10", "E10"]
for cell in center_cells:
    ws[cell].alignment = Alignment(horizontal='center')

# ファイル保存
wb.save("請求書_20251002.xlsx")
print("請求書_20251002.xlsx を保存しました。")
```
